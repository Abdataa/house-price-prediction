```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error,mean_squared_error
from sklearn.compose import ColumnTransformer
```

```python
target='SalePrice'
y_train=df[target]
X_train=df.drop(target
                ,axis=1)
```

```python
#feat_eng=['TotalSF']
#X_train.drop(feat_eng,axis=1,inplace=True)
```

```python
#featuring engineering practice 
from sklearn.base import BaseEstimator, TransformerMixin
class featureAdder(BaseEstimator, TransformerMixin):
    def __init__(self,drop_feature=None):
        self.drop_feature=drop_feature
        
    def fit(self,x,y=None):
        return self

        
    def transform(self,x):
        x_new=x.copy()
        x_new["TotalFlrSF"]=x_new["1stFlrSF"]+x_new["2ndFlrSF"]
        x_new["BsmtFinSFT"]=x_new['BsmtFinSF2']+x_new['BsmtFinSF1']
        x_new['age']=x_new['YrSold']+x_new['YearBuilt']
        if self.drop_feature :
            x_new.drop(self.drop_feature,axis=1,inplace=True)
        return x_new
```

```python
num_preprocessing=Pipeline([
                           ('num_imputing',SimpleImputer(strategy='mean')),
                            ('scaling_num',StandardScaler()),
                              ])
cat_preprocessing=Pipeline([('cat_imputing',SimpleImputer(strategy='most_frequent')),
                          ('ohe_cat',OneHotEncoder(handle_unknown='ignore'))])
```

```python
num_preprocessing
```

```python
cat_preprocessing
```

```python
cat_attributes=list(X_train.select_dtypes(include="object").columns)
```

```python

prepared_feat=ColumnTransformer([
                                ('num_attr',num_preprocessing,num_attributes),
                                    ('cat_attr',cat_preprocessing,cat_attributes)
                                  ])
```

```python
model_pipe=Pipeline(steps=[('preprocessing',prepared_feat),                           
                           ('model',LinearRegression())])
```

```python
model_pipe
```

```python
model_pipe.fit(X_train,y_train)
```

```python
rmse=root_mean_squared_error(y_train,model_pipe.predict(X_train))
rmse
```

```python
residual=y_train-model_pipe.predict(X_train)
```

```python
residual.plot(kind='hist',bins=55,figsize=(10,6))# visualinzing model's residual 
plt.title('Residual of the Model')# 
plt.show()
```

```python
plt.scatter(x=residual,y=model_pipe.predict(X_train))
plt.title('Residual vs Predicted Values')
plt.show()
```

```python
model=model_pipe.named_steps['model']
coef=model.coef_*(model.coef_.std())
coef[:10]
```

```python
feat_name=model_pipe.named_steps['preprocessing'].get_feature_names_out()#toimport features that-
                                                                      # used for training
```

```python
coef_df=pd.DataFrame({'feat_name':feat_name,'coeficient':model.coef_}) # to features in dataframe
coef_df.sort_values(by='coeficient',ascending=False)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_pipe=Pipeline(steps=[('preprocessing',prepared_feat),#applying preprocessing
                              ('decision_tree',DecisionTreeRegressor())])#model initializing
```

```python
tree_pipe.fit(X_train,y_train)
```

```python
tree_rmse=root_mean_squared_error(y_train,tree_pipe.predict(X_train))
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation
scores = cross_val_score(model_pipe, X_train, y_train, scoring='neg_mean_squared_error', cv=5)

# Convert negative MSE scores to RMSE
rmse_scores = np.sqrt(-scores)

# Print the mean and standard deviation of RMSE
def cal_score(rmse):
    
    print("RMSE Scores:", rmse)
    print("Mean RMSE:", rmse.mean())
    print("Standard Deviation RMSE:", rmse.std())
cal_score(rmse_scores)    
```

```python
scores = cross_val_score(tree_pipe, X_train, y_train, scoring='neg_mean_squared_error', cv=5)

# Convert negative MSE scores to RMSE
rmse_scores = np.sqrt(-scores)

# Print the mean and standard deviation of RMSE
def cal_score(rmse):
    
    print("RMSE Scores:", rmse)
    print("Mean RMSE:", rmse.mean())
    print("Standard Deviation RMSE:", rmse.std())
cal_score(rmse_scores)  
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg=Pipeline(steps=[('preprocessing',prepared_feat), # first apply preprocessing 
                              ('decision_tree',RandomForestRegressor())])# model initializing
forest_reg
```

```python
forest_reg.fit(X_train,y_train) # fitting data to the model
```

```python
forest_rms=root_mean_squared_error(y_train,forest_reg.predict(X_train))# calculate metrice to 
                                                            #check model perfomance
forest_rms
```

```python
from sklearn.model_selection import cross_val_score
forest_mse_score=cross_val_score(forest_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)#evaluating
forest_mse_score=np.sqrt(-forest_mse_score)##model performance by cross_val_score
cal_score(forest_mse_score)
```

```python
path='data//test.csv'
test_set=pd.read_csv(path)

test_pred=forest_reg.predict(test_set)
submission = pd.DataFrame({'Id': test_set['Id'], 'SalePrice': test_pred})
# Save to CSV
submission.to_csv('submission.csv', index=False)
```

```python
# hyperparameter tunnning GridSeachCV
from sklearn.model_selection import GridSearchCV
#defining grid parameters for tunning
param_grid = {
    'decision_tree__n_estimators': [10, 25, 50],
    'decision_tree__max_depth': [None, 5, 10, 15],
    'decision_tree__min_samples_split': [2, 5],
    'decision_tree__min_samples_leaf': [1, 2],
    'decision_tree__max_features': ['sqrt', 'log2'],  # Fixed 'auto'
    'decision_tree__bootstrap': [True, False]
}
# Create GridSearchCV object and fit it to the data
grid_search = GridSearchCV(estimator=forest_reg, param_grid=param_grid, cv=5, 
                           n_jobs=-1, verbose=2)
# Fitting the model with GridSearchCV
grid_search.fit(X_train, y_train)
# printing out best param
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: ", grid_search.best_score_)
```
